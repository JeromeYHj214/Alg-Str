## 程序计时

```python
import time
import datetime
# 测试函数运行时间
def cal_time(fn):
    """计算性能的修饰器"""
    def wrapper(*args,**kwargs):
        starTime = time.time()
        f = fn(*args,**kwargs)
        endTime = time.time()
        print('%s() runtime:%s ms' % (fn.__name__, 1000*(endTime - starTime)))
        return f
    return wrapper
```

## 二分法（列表需排好序）

### 时间复杂度O(logn)

```python
#li必须排好序
@cal_time #计时用
def binary_search(li,val):  
    left = 0
    right = len(li) - 1
    while left <= right:
        mid = (left + right) // 2
        if val == li[mid]:
            return mid
        elif val < li[mid]:
            right = mid - 1
        else:
            left = right + 1
    else:
        return None
```

```python
li = list(range(100000))
```

```python
binary_search(li,35410)
```

## 顺序查找

### 时间复杂度O(n)

```python
@cal_time #计时用
def linear_search(li,val):
    for ind, v in enumerate(li):
        if val == v:
            return ind
    else:
        return None
```

```python
linear_search(li,35410)
```

## 冒泡排序法

### 算法复杂度O(n^2)

```python
@cal_time
def bubble_sort(li):
    for i in range(len(li)-1):
        exchange = False
        for j in range(len(li)-i-1):
            if li[j] > li[j+1]:
                li[j+1], li[j] = li[j], li[j+1]
                exchange = True
        print(li)
        if exchange == False:
            return
```

### 若一次排序未发生变化，则整个列表排序完成

```python
li =[0,2,4,3,5,8,9]
```

```python
bubble_sort(li)
```

## 选择排序法

```python
@cal_time
def select_sort_simple(li):
    li_new = []
    for i in range(len(li)):
        min_val = min(li)
        li_new.append(min_val)
        li.remove(min_val)
    return li_new
```

```python
li =[0,2,4,3,5,8,9,3,6,4]
```

```python
select_sort_simple(li)
```

```python
@cal_time
def select_sort(li):
    for i in range(len(li)-1):
        min_loc = i
        for j in range(i+1,len(li)):
            if li[j] < li[min_loc]:
                min_loc = j
        li[i], li[min_loc] = li[min_loc], li[i]
        print(li)
```
