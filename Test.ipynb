## 程序计时

```python
import time
import datetime
# 测试函数运行时间
def cal_time(fn):
    """计算性能的修饰器"""
    def wrapper(*args,**kwargs):
        starTime = time.time()
        f = fn(*args,**kwargs)
        endTime = time.time()
        print('%s() runtime:%s ms' % (fn.__name__, 1000*(endTime - starTime)))
        return f
    return wrapper
```

## 汉诺塔问题（递归）

```python
def hanio(n,a,b,c):
    if n >0:
        hanio(n-1,a,c,b)
        print("moving from %s to %s"%(a,c))
        hanio(n-1,b,a,c)
```

# 查找问题

## 二分法（列表需排好序）时间复杂度O(logn)

```python
#li必须排好序
@cal_time #计时用
def binary_search(li,val):  
    left = 0
    right = len(li) - 1
    while left <= right:
        mid = (left + right) // 2
        if val == li[mid]:
            return mid
        elif val < li[mid]:
            right = mid - 1
        else:
            left = right + 1
    else:
        return None
```

```python
li = list(range(100000))
```

```python
binary_search(li,35410)
```

## 顺序查找 时间复杂度O(n)

```python
@cal_time #计时用
def linear_search(li,val):
    for ind, v in enumerate(li):
        if val == v:
            return ind
    else:
        return None
```

```python
linear_search(li,35410)
```

# 排序

## 冒泡排序法 算法复杂度O(n^2)

```python
@cal_time
def bubble_sort(li):
    for i in range(len(li)-1):
        exchange = False
        for j in range(len(li)-i-1):
            if li[j] > li[j+1]:
                li[j+1], li[j] = li[j], li[j+1]
                exchange = True
        #print(li)
        if exchange == False:
            return
```

### 若一次排序未发生变化，则整个列表排序完成

```python
li =[0,2,4,3,5,8,9]
```

```python
bubble_sort(li)
```

## 选择排序法 时间复杂度O(n^2)

```python
@cal_time
def select_sort_simple(li):
    li_new = []
    for i in range(len(li)):
        min_val = min(li)
        li_new.append(min_val)
        li.remove(min_val)
    return li_new
```

```python
li =[0,2,4,3,5,8,9,3,6,4]
```

```python
select_sort_simple(li)
```

## 插入排序法 时间复杂度O(n^2)

```python
@cal_time
def insert_sort(li):
    for i in range(1,len(li)):
        temp = li[i]
        j = i - 1
        while j >= 0 and li[j] > temp:
            li[j+1] = li[j]
            j -= 1
        li[j+1] = temp
```

```python
li =[0,2,4,3,5,8,9,3,6,4]
```

## 快速排序法 时间复杂度O(nlogn)

```python
import random
import copy
```

```python
def partition(data,left,right):
    #ind = random.randint(left,right)
    #data[left], data[ind] = data[ind], data[left] #防止最坏情况 data倒序排序 导致世间复杂度为O(n^2)
    temp = data[left]
    while left < right:
        while left < right and data[right] >= temp: #left > right是为了防止超出索引 使right为-1
            right -= 1
        data[left] = data[right]
        while left < right and data[left] <= temp:
            left += 1
        data[right] = data[left]
    data[left] = temp
    return left
```

```python
def _quick_sort(data,left,right):
    if left < right:
        mid = partition(data,left,right)
        _quick_sort(data,left,mid-1)
        _quick_sort(data,mid+1,right)
```

### 使用递归必须添加终止条件

```python
@cal_time
def quick_sort(data):
    _quick_sort(data,0,len(data)-1)
```

## 堆排序

```python
def sift(li,low,high):
    ############
    # li:列表
    # low：堆得根节点位置
    # high：堆得最后一个位置
    ###########
    i = low                  # i指向根节点
    j = 2*i + 1              # j指向孩子结点
    tmp = li[low]            #保存堆顶
    while j <= high:        #只要j节点有效
        if j + 1 <= high and li[j+1] > li [j]:  #右孩子存在且比较大
            j = j+1          #指向右孩子
        if tmp < li[j]:      
            li[i] = li[j]
            i = j            #往下一层
            j = 2*i + 1
        else:                #tmp值更大，放在位置i
            li[i] = tmp
            break
    else:
        li[i] = tmp         #tmp查到最底层，跳出while，tmp还未放回，所有加一步放回
            
```

### 当while因为条件跳出循环会去执行else，若因为break跳出循环，则不执行else。

```python
 def heap_sort(li):
        n = len(li)
        for i in range((n-2)//2, -1, -1):
            # i表示构建堆时的调整的堆下标
            sift(li, i, n-1)
        print(li)
```

```python
li = list(range(1,20))
print(li)
```

```python
heap_sort(li)
```
